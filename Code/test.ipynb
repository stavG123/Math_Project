```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pandas as pd  # Import pandas to read the file

# Load the Excel file
file_path = "../Excel/Data_Excel.xlsx"


# Read the three tables again
swimmers_df = pd.read_excel(file_path, sheet_name="Swimmer_ID")
performance_df = pd.read_excel(file_path, sheet_name="Bridge_table")
practice_df = pd.read_excel(file_path, sheet_name="PracticeDate", usecols=["Practice_ID", "Date"])

# Merge the data
merged_df = pd.merge(performance_df, swimmers_df, on="Swimmer_ID")
full_df = pd.merge(merged_df, practice_df, on="Practice_ID")

# Convert 'Date' column to datetime format
full_df["Date"] = pd.to_datetime(full_df["Date"])
# Define stroke categories based on max distance per practice
def get_dominant_stroke(row):
    strokes = {
        "Backstroke": row["Backstroke"],
        "Freestyle": row["Freestyle"],
        "Butterfly": row["Butterfly"],
        "Breaststroke": row["Breaststroke"],
        "Kick": row["Kick"]
    }
    return max(strokes, key=strokes.get)  # Get stroke with highest distance

# Apply function to create a new column in the DataFrame
full_df["Dominant_Stroke"] = full_df.apply(get_dominant_stroke, axis=1)

# Scatter plot: Calories Burned vs. Total Distance, color-coded by stroke
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=full_df["Total_Distance"],
    y=full_df["Calories_Burned"],
    hue=full_df["Dominant_Stroke"],  # Color by stroke type
    palette="Set2",  # Nice color scheme
    alpha=0.7  # Make points slightly transparent
)

# Formatting the graph
plt.xlabel("Total Distance (yards)")
plt.ylabel("Calories Burned (kcal)")
plt.title("Calories Burned vs. Total Distance (Color-Coded by Stroke)")
plt.legend(title="Dominant Stroke")
plt.grid(True)

# Show the plot
plt.show()


```
